# who is hightest wicket taker?

```python
bol=df[df['isWicketDelivery']==1]
```

```python
bol
```

```python
bol.groupby('bowler')['isWicketDelivery'].sum().sort_values(ascending=False)
```

```python
def wicket(bowler_name):
    bol=df[df['isWicketDelivery']==1]
    name=bol.groupby('bowler')
    name['bowler']==bowler_name
    return name['isWicketDelivery'].sum().sort_values(ascending=False)
```

```python
wicket('DJ Bravo')
```

```python
dm=df[df['bowler']=='HH Pandya']
```

```python
dm.groupby('BattingTeam')['isWicketDelivery'].sum().sort_values(ascending=False)
```

```python
def highest_wicket(bowler_name):
    dm=df[df['bowler']==bowler_name]
    return dm.groupby('BattingTeam')['isWicketDelivery'].sum().sort_values(ascending=False).head(3)
    
```

```python
 highest_wicket('DJ Bravo')
```
